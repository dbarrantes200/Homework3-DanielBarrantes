```python
# Dependencies and Setup
import pandas as pd
import numpy as np
%config IPCompleter.greedy = True

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
total_players = (purchase_data["SN"]).count()
total_players = {'Total Players': [total_players]
                 }
total_players_df = pd.DataFrame(total_players)
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Run basic calculations to obtain number of unique items, average price, etc.
unique_items = len(purchase_data["Item ID"].unique())
average = purchase_data["Price"].mean()
number_purchases = (purchase_data["Purchase ID"]).count()
total_revenue = purchase_data['Price'].sum()

  
total = {"Number of Unique Items": [unique_items],
         "Average Price":[average],
         "Number of Purchases": [number_purchases],
         "Total Revenue": [total_revenue]
         }
purchase_df = pd.DataFrame(total)
pd.options.display.float_format = "${:,.2f}".format
purchase_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
df = purchase_data.set_index('Gender')
males = df.loc['Male', 'SN'].count()
females = df.loc['Female', 'SN'].count()
other = df.loc['Other / Non-Disclosed', 'SN'].count()
p_male_players = (purchase_data['Gender'] == 'Male').mean()*100
p_female_players = (purchase_data['Gender'] == 'Female').mean()*100
p_other_players = (purchase_data['Gender'] == 'Other / Non-Disclosed').mean()*100


gender_demographics = {"Gender":["Male","Female","Other / Non-Disclosed"],
        "Total Count": [males, females, other],
        "Percentage of Players":[p_male_players, p_female_players, p_other_players]
        }
gender_demographics_df = pd.DataFrame(gender_demographics)
#pd.options.display.float_format = "${:,.2f}".format
gender_demographics_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
analysis_df = purchase_data[['Gender','SN','Price','Age']]
pd.options.display.float_format = "${:,.2f}".format
```

```python
purchasing_analysis_df = analysis_df.groupby(["Gender"])["SN"].agg([("Purchase Count","count")]).join(\
                    analysis_df.groupby(["Gender"])["Price"].agg([("Average Purchase Price","mean"),\
                                                                  ("Total Purchase Value","sum")]))
```

```python
purchasing_analysis_df["Avg Total Purchase per Person"] =purchasing_analysis_df["Total Purchase Value"] / \
                 analysis_df.groupby(["Gender"])["SN"].nunique()
```

* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points

```python
agedemo_df = purchase_data[['Age','SN']]
agedemo_df=agedemo_df.drop_duplicates()
#pd.options.display.float_format = "${:,.2f}".format
```

```python
bins = [0,9,14,19,24,29,34,40,99]
agedemo_df['Test Score Summary'] = pd.cut(agedemo_df['Age'], bins,labels=["<10", "10-14", "15-19", "20-24", "25-29","30-34", "35-39", "40+"])
#analysis1_df.head()
```

```python
grouped = agedemo_df.groupby('Test Score Summary').agg([('Counts','count')])
#grouped_mean = grouped.mean()

grouped['Percentage of Players']=round((grouped["Age"]/grouped["Age"].sum() * 100),2)
grouped

```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popularitems_df = purchase_data.copy()
popularitems_df.head()
pd.options.display.float_format = "${:,.2f}".format
```

```python
popularitems_df = popularitems_df.groupby(["Item ID", "Item Name"])["Purchase ID"].agg([("Purchase Count","count")]).join(\
                    popularitems_df.groupby(["Item ID", "Item Name"])["Price"].agg([("Item Price","min"),\
                                                                                     ("Total Purchase Value","sum")]))

popularitems_df.sort_values("Purchase Count", inplace=False, ascending=False).head(5)
```

```python
popularitems_df.sort_values("Total Purchase Value", inplace=False, ascending=False).head(5)
```
